```python
import pandas as pd
import glob
```

```python
path ='Data'
mex_filenames = glob.glob(path + "/*mex.csv")
usa_filenames = glob.glob(path + "/*usa.csv")
mex = []
usa = []
for filename in mex_filenames:
    mex.append((filename, pd.read_csv(filename)))
for filenames in usa_filenames:
    usa.append((filenames, pd.read_csv(filenames)))
```

```python
uac_usa = []
for i, v in usa:
    if i[5] == 'g':
        v['pais'] = 'Guatemala'            
    elif i[5] == 'e':
        v['pais'] = 'El Salvador'
    else:
        v['pais'] = 'Honduras'
    
    if i[-8] == '4':
        uac_usa.append(v[v['EDAD'] < 18])
    elif i[-8] == '5':
        uac_usa.append(v[v['edad'] < 18])
    elif i[-8] == '6':
        uac_usa.append(v[v['edad'] < 18])
    elif i[-8] == '7':
        uac_usa.append(v[v['edad'] < 18])
```
